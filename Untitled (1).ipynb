```python
# Book aur uske upar kitna rating para hai
ratings_with_books.groupby('title')['book-rating'].count()
```

```python
ratings_with_books.groupby('title')['book-rating'].count().reset_index()
```

```python
number_rating=ratings_with_books.groupby('title')['book-rating'].count().reset_index()
```

```python
number_rating
```

```python
number_rating.rename(columns={'book-rating':'No of rating'},inplace=True)
```

```python
number_rating
```

```python
final_rating=ratings_with_books.merge(number_rating,on='title')
```

```python
final_rating
```

```python
final_rating['No of rating']>=50
```

```python
final_rating=final_rating[final_rating['No of rating']>=50]
```

```python
final_rating
```

```python
final_rating.drop_duplicates(['user-iD','title'],inplace=True)
```

```python
final_rating.shape
```

```python
book_pivot=final_rating.pivot_table(columns='user-iD',index='title',values='book-rating')
```

```python
book_pivot
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse=csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
book_pivot.index==237
```

```python
distances,suggestions=model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1),n_neighbors=6)
```

```python
distances
```

```python
suggestions
```

```python
for i in range(len(suggestions)):
    print(book_pivot.index[suggestions[i]])
```
